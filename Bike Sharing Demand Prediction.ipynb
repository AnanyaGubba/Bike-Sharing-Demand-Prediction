```python
# Assigning string values to different seasons instead of numeric values

# 1 - SPRING
df.loc[(df['season'] == 1), 'season'] = 'spring'

# 2 - SUMMER
df.loc[(df['season'] == 2), 'season'] = 'summer'

# 3 - FALL
df.loc[(df['season'] == 3), 'season'] = 'fall'

# 4 - WINTER
df.loc[(df['season'] == 4), 'season'] = 'winter'
```

```python
df.tail()
```

```python
df['season'].astype('category').value_counts()
```

```python
# 0 = 2018
# 1 = 2019
df['yr'].astype('category').value_counts()
```

```python
# Assigning string values to different monthsll instead of numeric values

def mnt_map(x):
    return x.map({1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'})
```

```python
df[['mnth']] = df[['mnth']].apply(mnt_map)
```

```python
df['mnth'].astype('category').value_counts()
```

```python
df['holiday'].astype('category').value_counts()
```

```python
def str_map_weekday(x):
    return x.map({1:'Mon', 2:'Tues', 3:'Wed', 4:'Thurs', 5:'Fri', 6:'Sat', 0:'Sun'})
```

```python
df[['weekday']] = df[['weekday']].apply(str_map_weekday)
```

```python
df['weekday'].astype('category').value_counts()
```

```python
df['workingday'].astype('category').value_counts()
```

```python
# 1 -> Clear, few clouds, partly cloudy
df.loc[(df['weathersit'] == 1), 'weathersit'] = 'A'

# 2 -> Mist, Cloudy
df.loc[(df['weathersit'] == 2), 'weathersit'] = 'B'

# 3 -> Light snow, Heavy rain
df.loc[(df['weathersit'] == 3), 'weathersit'] = 'C'
```

```python
df['weathersit'].astype('category').value_counts()
```

# 3. Data Preparation

```python
df_categorical = df.select_dtypes(include='object')
```

```python
df_categorical.head()
```

```python
df_dummies = pd.get_dummies(df_categorical, drop_first=True)
df_dummies = df_dummies.astype('int')
```

```python
df_dummies.head()
```

```python
# Drop Categorical variable columns
df = df.drop(list(df_categorical.columns), axis=1)
df
```

```python
# Concatenate dummy variables with the dataset
df = pd.concat([df, df_dummies], axis=1)
```

```python
df.head()
```

```python
df = df.drop(['instant','dteday'], axis=1, inplace= False)
df.head()
```

# 4. Model Building and Evaluation

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
# Split the dataframe into training and test set

from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size=0.7, test_size=0.3, random_state=100)
```

```python
df_train
```

```python
df_train.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
var = ["temp","hum","windspeed","casual","registered","cnt"]

df_train[var] = scaler.fit_transform(df_train[var])
```

```python
df_train.describe().T
```

```python
# Correlation heatmap to know which variables are highly correlated

plt.figure(figsize=(30,30))
sns.heatmap(df_train.corr(), linewidth=0.5, annot=True, cmap="mako")
plt.title("Correlation Heatmap", fontsize=20)
plt.show()
```

```python
y_train = df_train.pop("cnt")
X_train = df_train.drop(["casual","registered"],axis=1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
np.asarray(df_train)
```

```python
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr = sm.OLS(y_train,X_train_lm).fit()
```

```python
X_train_lm
```

```python
lr.params
```

```python
lm = LinearRegression()

lm.fit(X_train_lm, y_train)
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
lr.summary()
```

```python
from sklearn.feature_selection import RFE
```

```python
lm = LinearRegression()
rfe1 = RFE(estimator=lm, n_features_to_select=15)

rfe1.fit(X_train,y_train)
print(rfe1.support_)
print(rfe1.ranking_)
```

```python
col1 = X_train.columns[rfe1.support_]
```

```python
col1
```

```python
X_train_rfe1 = X_train[col1]

X_train_rfe1 = sm.add_constant(X_train_rfe1)
lm1 = sm.OLS(y_train, X_train_rfe1).fit()
lm1.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
a = X_train_rfe1.drop('const',axis = 1)
```

```python
vif = pd.DataFrame()
vif['features'] = a.columns
vif['VIF'] = [variance_inflation_factor(a.values, i) for i in range(a.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
lm = LinearRegression()
rfe2 = RFE(estimator=lm, n_features_to_select=7)

rfe2.fit(X_train,y_train)
print(rfe2.support_)
print(rfe2.ranking_)
```

```python
col2 = X_train.columns[rfe2.support_]

X_train_rfe2 = X_train[col2]

X_train_rfe2 = sm.add_constant(X_train_rfe2)
lm2 = sm.OLS(y_train,X_train_rfe2).fit()
lm2.summary()
```

```python
b = X_train_rfe2.drop(['const'],axis=1)
vif1 = pd.DataFrame()
vif1['features'] = b.columns
vif1['VIF'] = [variance_inflation_factor(b.values, i) for i in range(b.shape[1])]
vif1['VIF'] = round(vif1['VIF'], 2)
vif1 = vif1.sort_values(by = "VIF", ascending = False)
vif1
```

```python
y_train_cnt = lm2.predict(X_train_rfe2)
```

```python
fig = plt.figure()
sns.distplot((y_train,y_train_cnt),bins=20)
```
